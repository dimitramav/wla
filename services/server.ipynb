```python
import nest_asyncio, uvicorn, os
nest_asyncio.apply()

port = int(os.getenv("RAG_PORT", "8000"))
config = uvicorn.Config("services.rag.api:app", host="0.0.0.0", port=port, reload=False, log_level="info")
server = uvicorn.Server(config)
await server.serve()
```

```python
import nest_asyncio, uvicorn, os
from api.main import create_app  # Ensure api.py is in the same directory or PYTHONPATH

app = create_app()
nest_asyncio.apply()

port = int(os.getenv("RAG_PORT", "8000"))
config = uvicorn.Config(app, host="0.0.0.0", port=port, reload=False, log_level="info")
server = uvicorn.Server(config)
print("Server ready on port", port)
```

```python
await server.serve()
```

```python
import sys, subprocess

print("Kernel Python:", sys.executable)

# Install CPU-only PyTorch first (avoids pulling CUDA deps)
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q",
    "torch", "--index-url", "https://download.pytorch.org/whl/cpu"
])

# Install project requirements
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q", "-r", "./requirements.txt"
])

# Jupyter widget support
subprocess.check_call([
    sys.executable, "-m", "pip", "install", "-q", "ipywidgets"
])

print("All packages installed.")
```

```python
import importlib

packages = [
    "fastapi", "uvicorn", "nest_asyncio", "dotenv", "docling",
    "chromadb", "sentence_transformers", "langchain_text_splitters",
    "orjson", "numpy", "pymongo", "pydantic_settings", "torch"
]

all_ok = True
for pkg in packages:
    try:
        importlib.import_module(pkg.replace("-", "_"))
        print(f"  {pkg}")
    except ImportError as e:
        print(f"  {pkg}: {e}")
        all_ok = False

print("\nAll packages ready!" if all_ok else "\nSome packages failed — check errors above.")
```

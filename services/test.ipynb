```python
from rag.vecstore import collection_for
topic = "school_anxiety"
docset_hash = "9fcad20bf991db740f651415d9729aeea5e50825ed5e33962983dc33b429f7e7"
col = collection_for(f"{topic}")

# Total entries in the collection (any version)
print("TOTAL:", col.count())

# Entries specifically for this topic+hash
records = col.get(
    where={"$and": [{"topic": topic}, {"docset_hash": docset_hash}]},
    include=["metadatas"]
)
print("records:", records)
print("MATCHED:", len(records.get("metadatas", [])))
# Peek first meta (should show 'topic' and 'docset_hash' keys)
metadatas = records.get("metadatas", [])
if metadatas:
    print(metadatas[0])
else:
    print("No metadata found.")
```
